```python
# Import các thư viện cần thiết
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import pandas as pd
from pathlib import Path
import platform

# Định nghĩa các hằng số
IMG_SIZE = (224, 224)  # Kích thước ảnh đầu vào cho EfficientNetB0
BATCH_SIZE = 32        # Kích thước batch
NUM_CLASSES = 29       # Số lớp (29 ký tự trong ASL)
EPOCHS = 15            # Số epoch huấn luyện
TRAIN_DIR = Path('../dataset/asl_dataset/train')  # Đường dẫn tới thư mục train
TEST_DIR = Path('../dataset/asl_dataset/test')    # Đường dẫn tới thư mục test
```

```python
def print_system_info():
    """Hiển thị thông tin hệ thống và phiên bản TensorFlow."""
    print("=" * 50)
    print(f"Hệ điều hành: {platform.system()} {platform.release()}")
    print(f"Phiên bản TensorFlow: {tf.__version__}")
    print(f"GPU có sẵn: {len(tf.config.list_physical_devices('GPU')) > 0}")
    print("=" * 50)

# Chạy hàm để kiểm tra hệ thống
print_system_info()
```

```python
def analyze_dataset(directory):
    """Phân tích số lượng ảnh trong mỗi lớp của dataset."""
    counts = {class_name: len(os.listdir(directory / class_name))
              for class_name in os.listdir(directory)
              if (directory / class_name).is_dir()}
    return counts

# Phân tích tập train
train_counts = analyze_dataset(TRAIN_DIR)
print("\nThống kê tập train:")
print(f"- Tổng số lớp: {len(train_counts)}")
print(f"- Tổng số ảnh: {sum(train_counts.values())}")
```

```python
def prepare_data():
    """Chuẩn bị dữ liệu huấn luyện, validation và test."""
    # Tạo ImageDataGenerator cho tập train và validation với các phép biến đổi ảnh
    train_datagen = ImageDataGenerator(
        rescale=1./255,           # Chuẩn hóa giá trị pixel về [0, 1]
        rotation_range=15,        # Xoay ảnh ngẫu nhiên tối đa 15 độ
        width_shift_range=0.1,    # Dịch ngang ngẫu nhiên
        height_shift_range=0.1,   # Dịch dọc ngẫu nhiên
        shear_range=0.1,          # Biến đổi cắt ngẫu nhiên
        zoom_range=0.1,           # Phóng to/thu nhỏ ngẫu nhiên
        fill_mode='nearest',      # Điền pixel bị trống bằng giá trị gần nhất
        validation_split=0.2      # Chia 20% tập train làm validation
    )

    # Tạo generator cho tập train
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',  # One-hot encoding cho nhãn
        subset='training',         # Lấy phần training
        seed=42                    # Đảm bảo tái lập
    )

    # Tạo generator cho tập validation
    val_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation',       # Lấy phần validation
        seed=42
    )

    # Chuẩn bị tập test từ danh sách file
    test_files = [f for f in os.listdir(TEST_DIR) if f.endswith('.jpg')]
    test_df = pd.DataFrame({
        'filename': test_files,
        'class': [f.split('_test')[0] for f in test_files]  # Trích xuất nhãn từ tên file
    })

    # Tạo ImageDataGenerator cho tập test (chỉ chuẩn hóa, không biến đổi)
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_dataframe(
        test_df,
        directory=TEST_DIR,
        x_col='filename',
        y_col='class',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False,             # Không xáo trộn để đánh giá chính xác
        classes=list(train_generator.class_indices.keys())  # Đảm bảo nhãn đồng nhất
    )

    return train_generator, val_generator, test_generator, test_df

# Chạy hàm để chuẩn bị dữ liệu
train_generator, val_generator, test_generator, test_df = prepare_data()
print(f"\nThống kê tập test:\n- Tổng số ảnh: {len(test_df)}")
```

```python
def build_model():
    """Xây dựng mô hình EfficientNetB0 với các lớp tùy chỉnh."""
    # Tải EfficientNetB0 đã được huấn luyện sẵn từ ImageNet
    base_model = EfficientNetB0(
        input_shape=(*IMG_SIZE, 3),  # Đầu vào: ảnh RGB 224x224
        include_top=False,           # Loại bỏ lớp fully connected mặc định
        weights='imagenet',          # Sử dụng trọng số pre-trained
        pooling='avg'                # Global average pooling để giảm chiều
    )
    base_model.trainable = False    # Đóng băng các tầng pre-trained

    # Xây dựng mô hình với các lớp tùy chỉnh
    model = Sequential([
        base_model,
        Dense(256, activation='relu'),  # Lớp fully connected với 256 neuron
        Dropout(0.5),                   # Dropout 50% để giảm overfitting
        Dense(NUM_CLASSES, activation='softmax')  # Lớp đầu ra với softmax
    ])

    # Compile mô hình
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),  # Optimizer Adam
        loss='categorical_crossentropy',                         # Loss cho phân loại đa lớp
        metrics=['accuracy']                                     # Đánh giá bằng accuracy
    )

    return model

# Khởi tạo và hiển thị cấu trúc mô hình
model = build_model()
model.summary()
```

```python
# Định nghĩa callbacks để cải thiện quá trình huấn luyện
callbacks = [
    EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True),  # Dừng sớm nếu val_loss không cải thiện
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=2, verbose=1)             # Giảm learning rate nếu val_loss không cải thiện
]

# Huấn luyện mô hình
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1  # Hiển thị tiến trình huấn luyện
)
```

```python
def plot_training_history(history):
    """Vẽ biểu đồ độ chính xác và loss trong quá trình huấn luyện."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Biểu đồ accuracy
    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title('Độ chính xác (Accuracy)', fontsize=14)
    ax1.set_xlabel('Epoch', fontsize=12)
    ax1.legend()

    # Biểu đồ loss
    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Loss', fontsize=14)
    ax2.set_xlabel('Epoch', fontsize=12)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# Vẽ biểu đồ
plot_training_history(history)
```

```python
# Đánh giá mô hình trên tập test
loss, acc = model.evaluate(test_generator)
print(f"\nKẾT QUẢ ĐÁNH GIÁ:")
print(f"Độ chính xác trên tập test: {acc*100:.2f}%")
print(f"Loss trên tập test: {loss:.4f}")
```

```python
def predict_asl(model, image_path, class_names):
    """Dự đoán ký hiệu ASL từ một ảnh."""
    # Đọc và xử lý ảnh
    img = cv2.imread(str(image_path))
    display_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Chuyển sang RGB để hiển thị
    img = cv2.resize(img, IMG_SIZE) / 255.0            # Resize và chuẩn hóa
    img = np.expand_dims(img, axis=0)                   # Thêm chiều batch

    # Dự đoán
    pred = model.predict(img)
    class_idx = np.argmax(pred)                         # Lấy chỉ số lớp dự đoán
    confidence = np.max(pred)                           # Độ tin cậy
    class_label = class_names[class_idx]                # Nhãn lớp

    # Hiển thị ảnh và kết quả
    plt.figure(figsize=(6, 6))
    plt.imshow(display_img)
    plt.axis('off')
    plt.title(f"Kết quả: {class_label}\nĐộ tin cậy: {confidence:.2%}", fontsize=14)
    plt.show()

# Kiểm tra ngẫu nhiên 3 ảnh từ tập test
class_names = list(train_generator.class_indices.keys())
print("\nKIỂM TRA NGẪU NHIÊN 3 ẢNH:")
for file in test_df.sample(3)['filename']:
    print("\n" + "=" * 50)
    print(f"File: {file}")
    predict_asl(model, TEST_DIR / file, class_names)
```

```python
# Lưu mô hình đã huấn luyện
model.save('asl_efficientnetb0_model.h5')
print("Đã lưu mô hình thành công!")
```
