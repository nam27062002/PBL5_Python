Import thư viện

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import pandas as pd
import platform
from pathlib import Path
import json
import seaborn as sns

# Định nghĩa các hằng số cơ bản cho quá trình huấn luyện
IMG_SIZE = (224, 224)  # Kích thước ảnh đầu vào (224x224 pixel)
BATCH_SIZE = 32        # Số lượng ảnh trong mỗi batch
NUM_CLASSES = 29       # Số lớp (tương ứng 29 ký tự ASL)
EPOCHS = 15            # Số lần huấn luyện tối đa
TRAIN_DIR = Path('../dataset/asl_dataset/train')  # Đường dẫn đến thư mục train
TEST_DIR = Path('../dataset/asl_dataset/test')    # Đường dẫn đến thư mục test
```

Hiển thị thông tin hệ thống

```python
def print_system_info():
    """In thông tin về hệ thống và môi trường thực thi."""
    print("=" * 50)
    print(f"Hệ điều hành: {platform.system()} {platform.release()}")
    print(f"Phiên bản TensorFlow: {tf.__version__}")
    print(f"Có GPU: {len(tf.config.list_physical_devices('GPU')) > 0}")
    print("=" * 50)

# Gọi hàm để hiển thị thông tin
print_system_info()
```

Phân tích dataset và vẽ biểu đồ phân bố

```python
def analyze_dataset(directory):
    """Đếm số lượng ảnh trong mỗi lớp của dataset."""
    counts = {class_name: len(os.listdir(directory / class_name))
              for class_name in os.listdir(directory)
              if (directory / class_name).is_dir()}
    return counts

# Phân tích tập train
train_counts = analyze_dataset(TRAIN_DIR)
print("\nThống kê tập train:")
print(f"- Tổng số lớp: {len(train_counts)}")
print(f"- Tổng số ảnh: {sum(train_counts.values())}")

# Vẽ biểu đồ phân bố số lượng ảnh theo lớp
plt.figure(figsize=(12, 6))
sns.barplot(x=list(train_counts.keys()), y=list(train_counts.values()))
plt.title('Phân bố số lượng ảnh theo lớp trong tập train', fontsize=16)
plt.xlabel('Lớp', fontsize=12)
plt.ylabel('Số lượng ảnh', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

Chuẩn bị dữ liệu huấn luyện, validation và test

```python
def prepare_data():
    """Chuẩn bị dữ liệu huấn luyện, validation và test."""
    # Khởi tạo ImageDataGenerator với các tham số tăng cường dữ liệu
    train_datagen = ImageDataGenerator(
        rescale=1./255,           # Chuẩn hóa giá trị pixel về [0, 1]
        rotation_range=15,        # Xoay ảnh ngẫu nhiên trong khoảng 15 độ
        width_shift_range=0.1,    # Dịch chuyển ngang
        height_shift_range=0.1,   # Dịch chuyển dọc
        shear_range=0.1,          # Biến dạng ảnh
        zoom_range=0.1,           # Phóng to/thu nhỏ
        fill_mode='nearest',      # Điền pixel bị thiếu
        validation_split=0.2      # Chia 20% dữ liệu cho validation
    )

    # Tạo generator cho tập train
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='training',
        seed=42
    )

    # Tạo generator cho tập validation
    val_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        subset='validation',
        seed=42
    )

    # Chuẩn bị dữ liệu test từ danh sách file
    test_files = [f for f in os.listdir(TEST_DIR) if f.endswith('.jpg')]
    test_df = pd.DataFrame({
        'filename': test_files,
        'class': [f.split('_test')[0] for f in test_files]
    })

    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_dataframe(
        test_df,
        directory=TEST_DIR,
        x_col='filename',
        y_col='class',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False,
        classes=list(train_generator.class_indices.keys())
    )

    # Lấy danh sách tên lớp
    class_names = list(train_generator.class_indices.keys())
    print(f"\nDanh sách các lớp: {class_names}")

    return train_generator, val_generator, test_generator, test_df, class_names

# Gọi hàm để chuẩn bị dữ liệu
train_generator, val_generator, test_generator, test_df, class_names = prepare_data()
print(f"\nThống kê tập test:\n- Tổng số ảnh: {len(test_df)}")
```

Xây dựng mô hình MobileNetV2

```python
def build_model():
    """Xây dựng mô hình MobileNetV2 với các lớp tùy chỉnh."""
    # Tải mô hình MobileNetV2 pre-trained
    base_model = MobileNetV2(
        input_shape=(*IMG_SIZE, 3),
        include_top=False,      # Loại bỏ lớp fully connected mặc định
        weights='imagenet',     # Sử dụng trọng số pre-trained từ ImageNet
        pooling='avg'           # Global average pooling
    )
    base_model.trainable = False  # Đóng băng các tầng pre-trained

    # Xây dựng mô hình tuần tự
    model = Sequential([
        base_model,
        Dense(256, activation='relu'),  # Lớp fully connected với 256 neuron
        Dropout(0.5),                   # Dropout để giảm overfitting
        Dense(NUM_CLASSES, activation='softmax')  # Lớp đầu ra với softmax
    ])

    # Biên dịch mô hình
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Khởi tạo và hiển thị cấu trúc mô hình
model = build_model()
model.summary()
```

Huấn luyện mô hình

```python
# Định nghĩa callbacks để điều chỉnh quá trình huấn luyện
callbacks = [
    EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=2, verbose=1)
]

# Huấn luyện mô hình
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1
)
```

Vẽ biểu đồ độ chính xác và loss

```python
def plot_training_history(history):
    """Vẽ biểu đồ độ chính xác và loss trong quá trình huấn luyện."""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Biểu đồ độ chính xác
    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Validation')
    ax1.set_title('Độ chính xác (Accuracy)', fontsize=14)
    ax1.set_xlabel('Epoch', fontsize=12)
    ax1.set_ylabel('Accuracy', fontsize=12)
    ax1.legend()

    # Biểu đồ loss
    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Validation')
    ax2.set_title('Loss', fontsize=14)
    ax2.set_xlabel('Epoch', fontsize=12)
    ax2.set_ylabel('Loss', fontsize=12)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# Gọi hàm để vẽ biểu đồ
plot_training_history(history)
```

Đánh giá mô hình trên tập test

```python
# Đánh giá mô hình trên tập test
loss, acc = model.evaluate(test_generator)
print(f"\nKẾT QUẢ ĐÁNH GIÁ:")
print(f"Độ chính xác trên tập test: {acc*100:.2f}%")
print(f"Loss trên tập test: {loss:.4f}")

# Dự đoán trên tập test
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes

# Vẽ confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix', fontsize=16)
plt.xlabel('Dự đoán', fontsize=12)
plt.ylabel('Thực tế', fontsize=12)
plt.show()

# In báo cáo chi tiết
import numpy as np
from sklearn.metrics import classification_report

# Lấy danh sách các lớp thực tế từ true_classes
unique_true_classes = np.unique(true_classes)
print(f"Số lớp thực tế trong tập test: {len(unique_true_classes)}")

# Lấy tên lớp tương ứng với các lớp thực tế
target_names = [class_names[i] for i in unique_true_classes]

# In báo cáo chi tiết với các lớp thực tế
print(classification_report(true_classes, predicted_classes, labels=unique_true_classes, target_names=target_names))
```

Lưu mô hình và danh sách lớp

```python
# Lưu mô hình
model.save('asl_mobile_net_v2_model.h5')

# Lưu danh sách class_names vào file JSON
with open('class_names.json', 'w') as f:
    json.dump(class_names, f)
print("Đã lưu mô hình và class_names thành công!")
```

Dự đoán và hiển thị kết quả cho một ảnh

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def predict_and_show(image_path, model, class_names):
    """Dự đoán và hiển thị kết quả cho một ảnh đơn lẻ."""
    # Tải và xử lý ảnh
    img = load_img(image_path, target_size=IMG_SIZE)
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Dự đoán
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)
    confidence = np.max(prediction)
    predicted_label = class_names[predicted_class]

    # Hiển thị ảnh và kết quả
    plt.imshow(img)
    plt.title(f'Dự đoán: {predicted_label} ({confidence:.2%})', fontsize=14)
    plt.axis('off')
    plt.show()

# Ví dụ: Dự đoán một ảnh trong tập test
image_path = TEST_DIR / test_df['filename'][0]
predict_and_show(image_path, model, class_names)
```
